```python
import openmc
import openmc.deplete
import numpy as np
import matplotlib.pyplot as plt
```

```python
nuc_list = ['U238', 'U235', 'U236', 'Pu238', 'Pu239', 'Pu240', 'Pu241']
```

```python
i0_res = openmc.deplete.Results("iter0/depletion_results.h5")
t, i0_k = i0_res.get_keff(time_units='d')
```

```python
i0_comp = {}

for n in nuc_list:
    i0_nuc = [i0_res.export_to_materials(i, 
                                         path='iter0/materials.xml', 
                                         nuc_with_data = nuc_list)[0].get_mass(nuclide = n) 
              for i in range(len(t))]
    i0_comp[n] = i0_nuc

i0_mass = np.sum([i0_comp[k][0] for k in i0_comp.keys()])
for k, v in i0_comp.items():
    i0_comp[k] = v/i0_mass
```

```python
for k, v in list(i0_comp.items())[1:]:
    plt.plot(t[1:], 100*v[1:], label = k)
plt.legend()
plt.ylabel("Isotopic Concentration [wt%]")
plt.xlabel("Time [days]")
plt.xlim((258, 1550))
plt.show()
```
